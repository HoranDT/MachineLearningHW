Exercise 2 (10 Points): As discussed in class, for the Iris dataset we shall use the sepal length
and the petal length as our choice of features. Write a Python program to draw, in the same
figure, a scatter plot for all feature vectors in the dataset. Mark each feature vector according to
its target label and make sure that the marks for the three different classes: setosa, versicolor, and
verginica, are markedly different. Based on your scatter plot, among the three pairs of classes:
setosa vs. versicolor, setosa vs. verginica, and versicolor vs. verginica, which pair(s) are linear
separable and which pair(s) are not linearly separable?

```python
import matplotlib.pyplot as plt                                                 #This library is used for plotting
import numpy as np                                                              #This library is used for numerical operations


y = df.iloc[:, 4].values                                                        #Selects the target labels for Setosa, Versicolor, and Virginica from the entire dataset (150 rows) and the 5th column (index 4)
y = np.where(y == 'Iris-setosa', 0, np.where(y == 'Iris-versicolor', 1, 2))     #Converts the target labels: 'Iris-setosa' becomes 0, 'Iris-versicolor' becomes 1, 'Iris-virginica' becomes 2


X = df.iloc[:, [0, 2]].values                                                   #Extracts the feature vectors for sepal length and petal length from the entire dataset (150 rows)
                                                                                #Sepal length is in the 1st column (index 0), and petal length is in the 3rd column (index 2)


plt.scatter(X[:50, 0], X[:50, 1],                                               #Creates a scatter plot for Setosa feature vectors (first 50 rows)
            color='red', marker='o', label='Setosa')                            #Red 'o' markers for Setosa were chosen

plt.scatter(X[50:100, 0], X[50:100, 1],                                         #Creates a scatter plot for Versicolor feature vectors (next 50 rows)
            color='blue', marker='s', label='Versicolor')                       #Blue 's' (stands for square) markers for Versicolor were chosen


plt.scatter(X[100:150, 0], X[100:150, 1],                                       #Creates a scatter plot for Virginica feature vectors (last 50 rows)
            color='green', marker='^', label='Virginica')                       #Use green '^' (stands for triangle) markers for Virginica

def plot_species_scatter(X, y, indices1, indices2, color1,                      #Defines a function to plot scatter plots for two species
                         marker1, color2, marker2, label1, label2):
    plt.figure()                                                                #Creates a new figure
    plt.scatter(X[indices1, 0], X[indices1, 1],                                 #Plots the first species using the specified indices, color, marker, and label
                color=color1, marker=marker1, label=label1)

    plt.scatter(X[indices2, 0], X[indices2, 1],                                 #Plots the second species using the specified indices, color, marker, and label
                color=color2, marker=marker2, label=label2)

plt.xlabel('Sepal length [cm]')                                                 #Labels the x-axis as 'Sepal length [cm]'
plt.ylabel('Petal length [cm]')                                                 #Labels the y-axis as 'Petal length [cm]'
plt.legend(loc='upper left')                                                    #Places the legend in the upper left corner of the plot
plt.tight_layout()                                                              #Adjusts the subplots of your figure so that they fit within the figure area in a "tight" manner
plt.show()                                                                      #Displays the scatter plot

plot_species_scatter(X, y, np.arange(0, 50), np.arange(50, 100), 'red', 'o',    #Plots for Setosa & Versicolor
                     'blue', 's', 'Setosa', 'Versicolor')                       #Uses indices 0-49 for Setosa and 50-99 for Versicolor
plt.xlabel('Sepal length [cm]')                                                 #Labels the x-axis as 'Sepal length [cm]'
plt.ylabel('Petal length [cm]')                                                 #Labels the y-axis as 'Petal length [cm]'
plt.legend(loc='upper left')                                                    #Places the legend in the upper left corner of the plot
plt.tight_layout()                                                              #Adjusts the subplots of your figure so that they fit within the figure area in a "tight" manner
plt.show()

plot_species_scatter(X, y, np.arange(50, 100), np.arange(100, 150), 'blue', 's',#Plots for Versicolor & Virginica
                     'green', '^', 'Versicolor', 'Virginica')                   #Uses indices 50-99 for Versicolor and 100-149 for Virginica
plt.xlabel('Sepal length [cm]')                                                 #Labels the x-axis as 'Sepal length [cm]'
plt.ylabel('Petal length [cm]')                                                 #Labels the y-axis as 'Petal length [cm]'
plt.legend(loc='upper left')                                                    #Places the legend in the upper left corner of the plot
plt.tight_layout()                                                              #Adjusts the subplots of your figure so that they fit within the figure area in a "tight" manner
plt.show()

plot_species_scatter(X, y, np.arange(0, 50), np.arange(100, 150), 'red', 'o',   #Plots for Setosa & Virginica
                     'green', '^', 'Setosa', 'Virginica')                       #Uses indices 0-49 for Setosa and 100-149 for Virginica
plt.xlabel('Sepal length [cm]')                                                 #Labels the x-axis as 'Sepal length [cm]'
plt.ylabel('Petal length [cm]')                                                 #Labels the y-axis as 'Petal length [cm]'
plt.legend(loc='upper left')                                                    #Places the legend in the upper left corner of the plot
plt.tight_layout()                                                              #Adjusts the subplots of your figure so that they fit within the figure area in a "tight" manner
plt.show()
```

Exercise 3 (20 Points): Write either a Python class or a Python function, which takes a set of
feature vectors, the corresponding binary labels, and the number of epochs as input, and output
an extended weight vector (which includes both the weights and the bias) based on the perceptron
learning algorithm that we discussed in the lecture. Test your program over all pairs of linearly
separable classes. For each pair of linearly separable classes, keep track of the update trajectory
of the extended weight vector by plotting, in two separate figures, the number of updates made
by the algorithm in each epoch and the number of errors made by the extended weight vector
produced at the conclusion of each epoch when tested against all training data examples. Plot
the decision boundary of the final output classifier.


```python
from matplotlib.colors import ListedColormap                                    #Imports colormap for plotting

                                                                                #Extracts labels and features for Setosa & Versicolor
y_sv = df.iloc[0:100, 4].values                                                 #Extracts labels for first 100 samples
y_sv = np.where(y_sv == 'Iris-setosa', 0, 1)                                    #Converts labels to 0 and 1
X_sv = df.iloc[0:100, [0, 2]].values                                            #Extracts features for first 100 samples

                                                                                #Extracts labels and features for Setosa & Virginica
y_svi = df.iloc[np.concatenate([np.arange(0, 50),                               #Extracts labels for Setosa & Virginica
                                np.arange(100, 150)]), 4].values
y_svi = np.where(y_svi == 'Iris-setosa', 0, 1)                                  #Converts labels to 0 and 1
X_svi = df.iloc[np.concatenate([np.arange(0, 50),                               #Extracts features for Setosa & Virginica
                                np.arange(100, 150)]), [0, 2]].values

class Perceptron:                                                               #Defines the Perceptron class
    def __init__(self, n_iter=50, random_state=1):                              #Initializes with number of iterations and random state
        self.n_iter = n_iter
        self.random_state = random_state
        self.updates_ = []                                                      #Lists to store the number of updates in each epoch

    def fit(self, X, y):                                                        #Fits the model
        self.w_ = np.zeros(X.shape[1])                                          #Initializes weights to zero
        self.b_ = np.float_(0.)                                                 #Initializes bias to zero
        self.extended_errors_ = []                                              #Lists to store the number of extended errors in each epoch
        for _ in range(self.n_iter):                                            #Loops over epochs
            updates = 0                                                         #Initializes updates counter
            for xi, target in zip(X, y):                                        #Loops over each sample
                update = 1 * (target - self.predict(xi))                        #Calculates the update
                self.w_ += update * xi                                          #Updates the weights
                self.b_ += update                                               #Updates the bias
                updates += int(update != 0.0)                                   #Increments updates counter if update is non-zero
            self.updates_.append(updates)                                       #Appends the number of updates for this epoch
            extended_errors = sum(self.predict(X) != y)                         #Calculates the number of extended errors
            self.extended_errors_.append(extended_errors)                       #Appends the number of extended errors for this epoch
        return self

    def net_input(self, X):                                                     #Calculates the net input
        return np.dot(X, self.w_) + self.b_

    def predict(self, X):                                                       #Makes predictions
        return np.where(self.net_input(X) >= 0.0, 1, 0)                         #Returns 1 if net_input >= 0 else return 0

                                                                                #Trains the Perceptron model for Setosa & Versicolor
ppn_sv = Perceptron(n_iter=10)                                                  #Initializes the Perceptron model
ppn_sv.fit(X_sv, y_sv)                                                          #Fits the model

                                                                                #Trains the Perceptron model for Setosa & Virginica
ppn_svi = Perceptron(n_iter=10)                                                 #Initializes the Perceptron model
ppn_svi.fit(X_svi, y_svi)                                                       #Fits the model

                                                                                #Plotting the number of updates for Setosa & Versicolor
plt.figure()                                                                    #Creates a new figure
plt.plot(range(1, len(ppn_sv.updates_) + 1), ppn_sv.updates_, marker='o')       #Plots the updates
plt.xlabel('Epochs')                                                            #Labels the x-axis
plt.ylabel('Number of Updates')                                                 #Labels the y-axis
plt.title('Number of Updates for Setosa & Versicolor')                          #Adds a title
plt.tight_layout()                                                              #Adjusts the layout
plt.show()                                                                      #Shows the plot

                                                                                #Plotting the number of extended errors for Setosa & Versicolor
plt.figure()                                                                    #Creates a new figure
plt.plot(range(1, len(ppn_sv.extended_errors_) + 1),                            #Plots the extended errors
         ppn_sv.extended_errors_, marker='x')
plt.xlabel('Epochs')                                                            #Labels the x-axis
plt.ylabel('Number of Extended Errors')                                         #Labels the y-axis
plt.title('Number of Extended Errors for Setosa & Versicolor')                  #Adds a title
plt.tight_layout()                                                              #Adjusts the layout
plt.show()                                                                      #Shows the plot

                                                                                #Plotting the number of updates for Setosa & Virginica
plt.figure()                                                                    #Creates a new figure
plt.plot(range(1, len(ppn_svi.updates_) + 1), ppn_svi.updates_, marker='o')     #Plots the updates
plt.xlabel('Epochs')                                                            #Labels the x-axis
plt.ylabel('Number of Updates')                                                 #Labels the y-axis
plt.title('Number of Updates for Setosa & Virginica')                           #Adds a title
plt.tight_layout()                                                              #Adjusts the layout
plt.show()                                                                      #Shows the plot

                                                                                #Plotting the number of extended errors for Setosa & Virginica
plt.figure()                                                                    #Creates a new figure
plt.plot(range(1, len(ppn_svi.extended_errors_) + 1),                           #Plots the extended errors
         ppn_svi.extended_errors_, marker='x')
plt.xlabel('Epochs')                                                            #Labels the x-axis
plt.ylabel('Number of Extended Errors')                                         #Labels the y-axis
plt.title('Number of Extended Errors for Setosa & Virginica')                   #Adds a title
plt.tight_layout()                                                              #Adjusts the layout
plt.show()                                                                      #Shows the plot

# Your existing code for plotting decision regions can be added here.
def plot_decision_regions(X, y, classifier, resolution=0.02):                   #Function to plot decision regions
    markers = ('o', 's', '^', 'v', '<')                                         #Defines markers for the plot
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')                      #Defines colors for the plot
    cmap = ListedColormap(colors[:len(np.unique(y))])                           #Creates a colormap based on the number of classes

    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1                       #Gets the min and max values for the two features
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),               #Creates a meshgrid based on the feature ranges
                           np.arange(x2_min, x2_max, resolution))

    lab = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)            #Makes predictions for each point in the meshgrid
    lab = lab.reshape(xx1.shape)

    plt.contourf(xx1, xx2, lab, alpha=0.3, cmap=cmap)                           #Plots the decision boundary
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):                                     #Plots the data points
        plt.scatter(x=X[y == cl, 0],
                    y=X[y == cl, 1],
                    alpha=0.8,
                    c=colors[idx],
                    marker=markers[idx],
                    label=f'Class {cl}',
                    edgecolor='black')

                                                                                #Plot decision regions for Setosa & Versicolor
plot_decision_regions(X_sv, y_sv, classifier=ppn_sv)
plt.xlabel('Sepal length [cm]')
plt.ylabel('Petal length [cm]')
plt.legend(loc='upper left')
plt.title('Decision Regions for Setosa & Versicolor')
plt.show()

                                                                                #Plot decision regions for Setosa & Virginica
plot_decision_regions(X_svi, y_svi, classifier=ppn_svi)
plt.xlabel('Sepal length [cm]')
plt.ylabel('Petal length [cm]')
plt.legend(loc='upper left')
plt.title('Decision Regions for Setosa & Virginica')
plt.show()
```

Exercise 4 (20 Points): Repeat Exercise 3 for the pair(s) of classes that are not linearly separable.
Does the algorithm converge in these cases? If not, analyze the update trajectories of the extended
weight vector and explain why it does not converge in these cases.

```python
                                                                                #Change the labels and features to include only "Iris-versicolor" and "Iris-virginica"
y = df.iloc[50:150, 4].values                                                   #Extracts the labels for the samples from index 50 to 149
y = np.where(y == 'Iris-versicolor', 1, 2)                                      #Converts the string labels to integers: 'Iris-versicolor' to 1 and 'Iris-virginica' to 2

X = df.iloc[50:150, [0, 2]].values                                              #Extracts the features for the samples from index 50 to 149

                                                                                #Run the perceptron algorithm
ppn = Perceptron(n_iter=50)                                                     #Creates a Perceptron object
ppn.fit(X, y)                                                                   #Fits the Perceptron model to the data

plt.figure()                                                                    #Creates a new figure for plotting
plt.plot(range(1, len(ppn.updates_) + 1), ppn.updates_, marker='o')             #Plots the number of updates (errors) in each epoch
plt.xlabel('Epochs')                                                            #Labels the x-axis as 'Epochs'
plt.ylabel('Number of Updates')                                                 #Labels the y-axis as 'Number of Updates'
plt.title('Update Trajectory for Non-linearly Separable Classes')               #Adds a title to the plot, specifying that this is the update trajectory for non-linearly separable classes
plt.tight_layout()                                                              #Adjusts the layout so that all elements fit well within the plot
plt.show()                                                                      #Displays the plot

plot_decision_regions(X, y, classifier=ppn)                                     #Calls the function to plot the decision regions
plt.xlabel('Sepal length [cm]')                                                 #Labels the x-axis as 'Sepal length [cm]'
plt.ylabel('Petal length [cm]')                                                 #Labels the y-axis as 'Petal length [cm]'
plt.legend(loc='upper left')                                                    #Adds a legend to the plot, specifying its location to be the upper left corner
plt.title('Decision Boundary for Non-linearly Separable Classes')               #Adds a title to the plot, specifying that this is the decision boundary for non-linearly separable classes
plt.show()                                                                      #Display the plot


plt.figure()
plt.plot(range(1, len(ppn.extended_errors_) + 1), ppn.extended_errors_, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Number of Extended Errors')
plt.title('Number of Extended Errors for Versicolor & Virginica')
plt.tight_layout()
plt.show()
```

By plotting graphs of two non-linearly separable classes, we can see that the trajectory line is making a constant number of updates, and the algorithm is not converging. The idea of a Perceptron is to find a hyperplane that separates the classes. For non-linearly separable classes, such a hyperplane doesn't exist. So, because of this, the algorithm keeps updating the weights in an attempt to minimize the error, leading to non-converging behavior.
